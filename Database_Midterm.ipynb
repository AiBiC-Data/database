```python
from sqlalchemy import create_engine
my_conn = create_engine('sqlite:////content/database_midterm.db')
```

```python
r_set=my_conn.execute('''
CREATE TABLE `Status` (
	`Vehicle Id`	VARCHAR(10)	NOT NULL,
	`Flag`	INT	NOT NULL,
	`Status`	VARCHAR(2)	NULL,
	`Date`	DATE  NULL,
	`Odometer`	INT	NULL,
	`Final Cost`	INT	NULL,
	`Description`	VARCHAR(1000)	NULL,
  PRIMARY KEY (`Vehicle Id`)
  FOREIGN KEY (`Vehicle Id`)
  REFERENCES `Can Use` (`Vehicle Id`),
  FOREIGN KEY (`Flag`)
  REFERENCES `Can Use` (`Flag`)
);
 ''')
```
